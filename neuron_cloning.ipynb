```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import os
from tqdm import tqdm
import glob
import gc

def optimize_gpu_memory():
    """Optimize GPU memory usage for Colab"""
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True  # Optimize for consistent input sizes
        torch.backends.cudnn.deterministic = False  # Allow non-deterministic for speed
        print("GPU memory optimized for Colab")

class IzhikevichDataset(Dataset):
    """Dataset class for Izhikevich neuron data"""

    def __init__(self, data_path, normalize=True, sequence_length=None):
        """
        Args:
            data_path: Path to directory containing batch files
            normalize: Whether to normalize the data
            sequence_length: Length of sequences to use (None = full length)
        """
        self.data_path = data_path
        self.normalize = normalize
        self.sequence_length = sequence_length

        # Load and process data
        self.load_data()
        if self.normalize:
            self.normalize_data()

    def load_data(self):
        """Load all batch files and combine them"""
        batch_files = sorted(glob.glob(os.path.join(self.data_path, "batch_*.npz")))

        all_v = []
        all_I = []
        all_params = []
        all_types = []

        print(f"Loading {len(batch_files)} batch files...")

        def safe_gpu_to_numpy(arr):
            """Safely convert GPU arrays to NumPy arrays"""
            if hasattr(arr, 'get'):  # CuPy array
                return arr.get()
            elif hasattr(arr, 'cpu'):  # PyTorch tensor
                return arr.cpu().numpy()
            elif hasattr(arr, 'numpy'):  # TensorFlow tensor
                return arr.numpy()
            else:
                return np.array(arr)

        for batch_file in tqdm(batch_files):
            try:
                data = np.load(batch_file, allow_pickle=True)

                # Extract data and ensure it's on CPU as NumPy arrays
                v_raw = data['v']
                I_raw = data['I']
                params_raw = data['params'].item()  # Dictionary of parameters
                types_raw = data['type_indices']

                # Convert GPU arrays to NumPy safely
                v = safe_gpu_to_numpy(v_raw).astype(np.float32)
                I = safe_gpu_to_numpy(I_raw).astype(np.float32)
                types = safe_gpu_to_numpy(types_raw)

                # Ensure I has same length as v by padding or truncating
                if I.shape[1] < v.shape[1]:
                    I_padded = np.zeros((I.shape[0], v.shape[1]), dtype=np.float32)
                    I_padded[:, :I.shape[1]] = I
                    I = I_padded
                elif I.shape[1] > v.shape[1]:
                    I = I[:, :v.shape[1]]

                all_v.append(v)
                all_I.append(I)
                all_types.append(types)

                # Stack parameters into array - ensure all are NumPy arrays
                param_arrays = []
                for param_name in ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']:
                    param_val = safe_gpu_to_numpy(params_raw[param_name])
                    param_arrays.append(param_val)

                param_array = np.stack(param_arrays, axis=1).astype(np.float32)
                all_params.append(param_array)

                # Clear memory
                data.close()
                del data, v_raw, I_raw, params_raw, types_raw

            except Exception as e:
                print(f"Error loading {batch_file}: {e}")
                import traceback
                traceback.print_exc()
                continue

        # Check if any data was loaded successfully
        if len(all_v) == 0:
            raise ValueError("No data could be loaded from any batch files. Check file paths and data format.")

        # Combine all batches - ensure everything is NumPy arrays
        print("Combining batches...")
        self.v_data = np.concatenate(all_v, axis=0).astype(np.float32)
        self.I_data = np.concatenate(all_I, axis=0).astype(np.float32)
        self.param_data = np.concatenate(all_params, axis=0).astype(np.float32)
        self.type_data = np.concatenate(all_types, axis=0)

        # Clear memory
        del all_v, all_I, all_params, all_types
        gc.collect()

        print(f"Loaded {self.v_data.shape[0]} neurons with {self.v_data.shape[1]} timesteps")

        # Truncate sequences if specified
        if self.sequence_length and self.sequence_length < self.v_data.shape[1]:
            self.v_data = self.v_data[:, :self.sequence_length]
            self.I_data = self.I_data[:, :self.sequence_length]

    def normalize_data(self):
        """Normalize input and output data"""
        # Ensure all data is NumPy arrays on CPU
        if hasattr(self.v_data, 'get'):  # CuPy array
            self.v_data = self.v_data.get()
        if hasattr(self.I_data, 'get'):  # CuPy array
            self.I_data = self.I_data.get()
        if hasattr(self.param_data, 'get'):  # CuPy array
            self.param_data = self.param_data.get()

        # Convert to NumPy if needed
        self.v_data = np.array(self.v_data).astype(np.float32)
        self.I_data = np.array(self.I_data).astype(np.float32)
        self.param_data = np.array(self.param_data).astype(np.float32)

        # Normalize voltage (convert to mV if in V)
        if np.max(np.abs(self.v_data)) < 1:  # Likely in volts
            self.v_data = self.v_data * 1000  # Convert to mV

        # Normalize sequences
        self.v_scaler = StandardScaler()
        self.I_scaler = StandardScaler()
        self.param_scaler = StandardScaler()

        # Reshape for scaling (flatten time dimension)
        v_flat = self.v_data.reshape(-1, 1)
        I_flat = self.I_data.reshape(-1, 1)

        # Fit scalers
        self.v_scaler.fit(v_flat)
        self.I_scaler.fit(I_flat)
        self.param_scaler.fit(self.param_data)

        # Transform data
        v_scaled = self.v_scaler.transform(v_flat).reshape(self.v_data.shape)
        I_scaled = self.I_scaler.transform(I_flat).reshape(self.I_data.shape)

        self.v_data = v_scaled.astype(np.float32)
        self.I_data = I_scaled.astype(np.float32)
        self.param_data = self.param_scaler.transform(self.param_data).astype(np.float32)

    def __len__(self):
        return len(self.v_data)

    def __getitem__(self, idx):
        # Stack v and I as input features
        features = np.stack([self.v_data[idx], self.I_data[idx]], axis=0)  # Shape: (2, seq_len)
        parameters = self.param_data[idx]  # Shape: (9,)

        return torch.FloatTensor(features), torch.FloatTensor(parameters)

class IzhikevichNet(nn.Module):
    """Neural network for predicting Izhikevich parameters from v and I traces"""

    def __init__(self, sequence_length, hidden_size=256, num_layers=3, dropout=0.2):
        super(IzhikevichNet, self).__init__()

        self.sequence_length = sequence_length
        self.hidden_size = hidden_size

        # CNN layers for feature extraction from time series
        self.conv_layers = nn.Sequential(
            nn.Conv1d(2, 32, kernel_size=5, padding=2),  # 2 input channels (v, I)
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Conv1d(32, 64, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.MaxPool1d(2),
            nn.Conv1d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(128),
            nn.MaxPool1d(2),
        )

        # Calculate conv output size
        conv_output_size = 128 * (sequence_length // 4)  # After 2 MaxPool1d layers

        # LSTM for temporal dependencies
        self.lstm = nn.LSTM(128, hidden_size, num_layers, batch_first=True, dropout=dropout)

        # Fully connected layers
        self.fc_layers = nn.Sequential(
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_size, hidden_size // 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_size // 2, 9)  # 9 parameters: C, k, vr, vt, vpeak, a, b, c, d
        )

    def forward(self, x):
        # x shape: (batch_size, 2, sequence_length)

        # CNN feature extraction
        conv_out = self.conv_layers(x)  # Shape: (batch_size, 128, seq_len//4)

        # Prepare for LSTM (batch_first=True)
        conv_out = conv_out.permute(0, 2, 1)  # Shape: (batch_size, seq_len//4, 128)

        # LSTM
        lstm_out, (h_n, c_n) = self.lstm(conv_out)

        # Use last hidden state
        last_hidden = h_n[-1]  # Shape: (batch_size, hidden_size)

        # Fully connected layers
        output = self.fc_layers(last_hidden)

        return output

class PhysicsInformedNet(nn.Module):
    """Physics-Informed Neural Network for Izhikevich parameters"""

    def __init__(self, sequence_length, hidden_size=256, num_layers=3, dropout=0.2, physics_weight=1.0):
        super(PhysicsInformedNet, self).__init__()

        self.physics_weight = physics_weight

        # Same architecture as regular network
        self.conv_layers = nn.Sequential(
            nn.Conv1d(2, 32, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Conv1d(32, 64, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.MaxPool1d(2),
            nn.Conv1d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(128),
            nn.MaxPool1d(2),
        )

        self.lstm = nn.LSTM(128, hidden_size, num_layers, batch_first=True, dropout=dropout)

        self.fc_layers = nn.Sequential(
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_size, hidden_size // 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_size // 2, 9)
        )

    def forward(self, x):
        conv_out = self.conv_layers(x)
        conv_out = conv_out.permute(0, 2, 1)
        lstm_out, (h_n, c_n) = self.lstm(conv_out)
        last_hidden = h_n[-1]
        output = self.fc_layers(last_hidden)
        return output

    def physics_loss(self, v, I, params, dt=0.05):
        """Calculate physics-based loss using Izhikevich equations"""
        # Unpack parameters
        C, k, vr, vt, vpeak, a, b, c, d = torch.chunk(params, 9, dim=1)

        # Calculate derivatives numerically
        dv_dt = torch.diff(v, dim=1) / dt  # Shape: (batch, seq_len-1)
        v_mid = v[:, :-1]  # Voltage at current timestep
        u_approx = a * (b * (v_mid - vr) - 0)  # Approximate u (simplified)
        I_mid = I[:, :-1]  # Current at current timestep

        # Izhikevich voltage equation: dv/dt = (k(v-vr)(v-vt) - u + I) / C
        dv_dt_predicted = (k * (v_mid - vr) * (v_mid - vt) - u_approx + I_mid) / C

        # Physics loss
        physics_loss = torch.mean((dv_dt - dv_dt_predicted) ** 2)

        return physics_loss

def train_model(model, train_loader, val_loader, device, epochs=100, learning_rate=1e-3,
                use_physics=False, physics_weight=1.0):
    """Train the neural network"""

    model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=10, factor=0.5)
    criterion = nn.MSELoss()

    train_losses = []
    val_losses = []

    best_val_loss = float('inf')
    patience_counter = 0
    patience = 20

    for epoch in range(epochs):
        # Training
        model.train()
        train_loss = 0.0
        physics_loss_total = 0.0

        for batch_idx, (features, targets) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}")):
            features, targets = features.to(device), targets.to(device)

            optimizer.zero_grad()

            outputs = model(features)
            loss = criterion(outputs, targets)

            # Add physics loss for PINN
            if use_physics and hasattr(model, 'physics_loss'):
                v_data = features[:, 0, :]  # Voltage data
                I_data = features[:, 1, :]  # Current data
                physics_loss = model.physics_loss(v_data, I_data, outputs)
                loss += physics_weight * physics_loss
                physics_loss_total += physics_loss.item()

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            train_loss += loss.item()

        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for features, targets in val_loader:
                features, targets = features.to(device), targets.to(device)
                outputs = model(features)
                loss = criterion(outputs, targets)
                val_loss += loss.item()

        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        physics_loss_total /= len(train_loader)

        train_losses.append(train_loss)
        val_losses.append(val_loss)

        print(f"Epoch {epoch+1}: Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}")
        if use_physics:
            print(f"Physics Loss: {physics_loss_total:.6f}")

        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping at epoch {epoch+1}")
                break

        scheduler.step(val_loss)

    return train_losses, val_losses

def evaluate_model(model, test_loader, param_scaler, device):
    """Evaluate the trained model"""
    model.eval()
    predictions = []
    true_values = []

    with torch.no_grad():
        for features, targets in test_loader:
            features, targets = features.to(device), targets.to(device)
            outputs = model(features)

            # Convert back to original scale
            outputs_cpu = outputs.cpu().numpy()
            targets_cpu = targets.cpu().numpy()

            if param_scaler:
                outputs_cpu = param_scaler.inverse_transform(outputs_cpu)
                targets_cpu = param_scaler.inverse_transform(targets_cpu)

            predictions.append(outputs_cpu)
            true_values.append(targets_cpu)

    predictions = np.concatenate(predictions, axis=0)
    true_values = np.concatenate(true_values, axis=0)

    # Calculate metrics
    mse = mean_squared_error(true_values, predictions)
    r2 = r2_score(true_values, predictions)

    param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']

    print(f"Overall MSE: {mse:.6f}")
    print(f"Overall R²: {r2:.6f}")

    # Per-parameter metrics
    for i, param in enumerate(param_names):
        param_mse = mean_squared_error(true_values[:, i], predictions[:, i])
        param_r2 = r2_score(true_values[:, i], predictions[:, i])
        print(f"{param}: MSE={param_mse:.6f}, R²={param_r2:.6f}")

    return predictions, true_values

def plot_results(predictions, true_values, param_names):
    """Plot prediction vs true values"""
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))
    axes = axes.flatten()

    for i, param in enumerate(param_names):
        ax = axes[i]
        ax.scatter(true_values[:, i], predictions[:, i], alpha=0.5)

        # Perfect prediction line
        min_val = min(true_values[:, i].min(), predictions[:, i].min())
        max_val = max(true_values[:, i].max(), predictions[:, i].max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.8)

        ax.set_xlabel(f'True {param}')
        ax.set_ylabel(f'Predicted {param}')
        ax.set_title(f'Parameter {param}')
        ax.grid(True, alpha=0.3)

        # Calculate R²
        r2 = r2_score(true_values[:, i], predictions[:, i])
        ax.text(0.05, 0.95, f'R² = {r2:.3f}', transform=ax.transAxes,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.savefig('parameter_predictions.png', dpi=150, bbox_inches='tight')
    plt.show()

# Main execution
if __name__ == "__main__":
    # Optimize for Colab
    optimize_gpu_memory()

    # Check and setup GPU
    if torch.cuda.is_available():
        device = torch.device('cuda')
        print(f"Using GPU: {torch.cuda.get_device_name(0)}")
        print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
        # Set GPU memory management
        torch.cuda.empty_cache()
    else:
        device = torch.device('cpu')
        print("GPU not available, using CPU")

    # Load dataset
    print("Loading dataset...")
    dataset = IzhikevichDataset("synthetic_dataset", sequence_length=1000)  # Use first 1000 timesteps

    # Split dataset
    train_size = int(0.7 * len(dataset))
    val_size = int(0.15 * len(dataset))
    test_size = len(dataset) - train_size - val_size

    print(f"Dataset split: Train={train_size}, Val={val_size}, Test={test_size}")

    train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(
        dataset, [train_size, val_size, test_size],
        generator=torch.Generator().manual_seed(42)  # For reproducibility
    )

    # Create data loaders - reduce num_workers for Colab
    batch_size = 64 if device.type == 'cuda' else 32  # Larger batch for GPU
    num_workers = 2 if device.type == 'cuda' else 0   # Reduce workers for Colab

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,
                             num_workers=num_workers, pin_memory=True if device.type == 'cuda' else False)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           num_workers=num_workers, pin_memory=True if device.type == 'cuda' else False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,
                            num_workers=num_workers, pin_memory=True if device.type == 'cuda' else False)

    # Train regular neural network
    print("\n" + "="*50)
    print("Training Regular Neural Network...")
    print("="*50)

    model = IzhikevichNet(sequence_length=1000, hidden_size=256)
    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")

    train_losses, val_losses = train_model(
        model, train_loader, val_loader, device,
        epochs=50, learning_rate=1e-3, use_physics=False
    )

    # Clear GPU cache
    if device.type == 'cuda':
        torch.cuda.empty_cache()

    print("\nEvaluating Regular Neural Network...")
    predictions, true_values = evaluate_model(
        model, test_loader, dataset.param_scaler if dataset.normalize else None, device
    )

    # Plot results
    param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']
    plot_results(predictions, true_values, param_names)

    # Save model
    torch.save(model.state_dict(), 'regular_nn_model.pth')
    print("Regular NN model saved as 'regular_nn_model.pth'")

    # Train PINN (optional)
    print("\n" + "="*50)
    print("Training Physics-Informed Neural Network...")
    print("="*50)

    pinn_model = PhysicsInformedNet(sequence_length=1000, hidden_size=256, physics_weight=1.0)
    print(f"PINN parameters: {sum(p.numel() for p in pinn_model.parameters()):,}")

    pinn_train_losses, pinn_val_losses = train_model(
        pinn_model, train_loader, val_loader, device,
        epochs=50, learning_rate=1e-3, use_physics=True, physics_weight=1.0
    )

    # Clear GPU cache
    if device.type == 'cuda':
        torch.cuda.empty_cache()

    print("\nEvaluating Physics-Informed Neural Network...")
    pinn_predictions, pinn_true_values = evaluate_model(
        pinn_model, test_loader, dataset.param_scaler if dataset.normalize else None, device
    )

    # Save PINN model
    torch.save(pinn_model.state_dict(), 'pinn_model.pth')
    print("PINN model saved as 'pinn_model.pth'")

    # Compare models
    print("\n" + "="*50)
    print("MODEL COMPARISON")
    print("="*50)

    # Calculate overall metrics for comparison
    regular_mse = mean_squared_error(true_values, predictions)
    regular_r2 = r2_score(true_values, predictions)
    pinn_mse = mean_squared_error(pinn_true_values, pinn_predictions)
    pinn_r2 = r2_score(pinn_true_values, pinn_predictions)

    print(f"Regular NN  - MSE: {regular_mse:.6f}, R²: {regular_r2:.6f}")
    print(f"PINN        - MSE: {pinn_mse:.6f}, R²: {pinn_r2:.6f}")

    if pinn_r2 > regular_r2:
        print("🏆 PINN performs better!")
    else:
        print("🏆 Regular NN performs better!")

    print("\n✅ Training complete!")
    print("📊 Check 'parameter_predictions.png' for results visualization")
    print("💾 Models saved for future use")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from tqdm import tqdm
import matplotlib.pyplot as plt

def optimize_gpu_memory():
    """Quick GPU optimization"""
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True
        print("GPU optimized")

class LightweightIzhikevichNet(nn.Module):
    """Lighter version of the neural network"""

    def __init__(self, sequence_length, hidden_size=128, dropout=0.3):
        super(LightweightIzhikevichNet, self).__init__()

        # Smaller CNN layers
        self.conv_layers = nn.Sequential(
            nn.Conv1d(2, 16, kernel_size=7, padding=3),
            nn.ReLU(),
            nn.BatchNorm1d(16),
            nn.Conv1d(16, 32, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.MaxPool1d(2),
            nn.Conv1d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.MaxPool1d(2),
            nn.Dropout1d(dropout),
        )

        # Smaller LSTM
        self.lstm = nn.LSTM(64, hidden_size, 2, batch_first=True, dropout=dropout)

        # Parameter-specific heads for better individual parameter prediction
        self.param_heads = nn.ModuleDict({
            'C': nn.Sequential(nn.Linear(hidden_size, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1)),
            'k': nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Dropout(dropout), nn.Linear(32, 1)),
            'vr': nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Dropout(dropout), nn.Linear(32, 1)),
            'vt': nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Dropout(dropout), nn.Linear(32, 1)),
            'vpeak': nn.Sequential(nn.Linear(hidden_size, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1)),  # More capacity for problematic param
            'a': nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Dropout(dropout), nn.Linear(32, 1)),
            'b': nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Dropout(dropout), nn.Linear(32, 1)),
            'c': nn.Sequential(nn.Linear(hidden_size, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1)),
            'd': nn.Sequential(nn.Linear(hidden_size, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1)),
        })

    def forward(self, x):
        # CNN feature extraction
        conv_out = self.conv_layers(x)
        conv_out = conv_out.permute(0, 2, 1)

        # LSTM
        lstm_out, (h_n, c_n) = self.lstm(conv_out)
        last_hidden = h_n[-1]

        # Parameter-specific predictions
        outputs = []
        param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']
        for param in param_names:
            outputs.append(self.param_heads[param](last_hidden))

        return torch.cat(outputs, dim=1)

def weighted_mse_loss(predictions, targets, weights=None):
    """Weighted MSE loss to focus on poorly performing parameters"""
    if weights is None:
        # Default weights - higher for parameters that performed poorly
        weights = torch.tensor([1.0, 1.0, 1.0, 1.0, 3.0, 1.0, 1.0, 2.0, 1.0])  # Higher weight for vpeak and c

    weights = weights.to(predictions.device)
    mse_per_param = torch.mean((predictions - targets) ** 2, dim=0)
    weighted_loss = torch.sum(weights * mse_per_param)
    return weighted_loss

def quick_train(model, train_loader, val_loader, device, epochs=15, lr=5e-4):
    """Fast training with optimizations"""
    model.to(device)

    # Different learning rates for different parts
    optimizer = optim.AdamW([
        {'params': model.conv_layers.parameters(), 'lr': lr * 0.5},
        {'params': model.lstm.parameters(), 'lr': lr},
        {'params': model.param_heads.parameters(), 'lr': lr * 2.0}  # Higher LR for heads
    ], weight_decay=1e-4)

    scheduler = optim.lr_scheduler.OneCycleLR(optimizer, max_lr=lr*2,
                                            steps_per_epoch=len(train_loader),
                                            epochs=epochs, pct_start=0.3)

    train_losses = []
    val_losses = []

    for epoch in range(epochs):
        # Training
        model.train()
        train_loss = 0.0

        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for features, targets in pbar:
            features, targets = features.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = model(features)

            # Use weighted loss
            loss = weighted_mse_loss(outputs, targets)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
            optimizer.step()
            scheduler.step()

            train_loss += loss.item()
            pbar.set_postfix({'loss': f'{loss.item():.4f}'})

        # Quick validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for features, targets in val_loader:
                features, targets = features.to(device), targets.to(device)
                outputs = model(features)
                loss = weighted_mse_loss(outputs, targets)
                val_loss += loss.item()

        train_loss /= len(train_loader)
        val_loss /= len(val_loader)

        train_losses.append(train_loss)
        val_losses.append(val_loss)

        print(f"Epoch {epoch+1}: Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

        # Save best model
        if epoch == 0 or val_loss < min(val_losses[:-1]):
            torch.save(model.state_dict(), 'lightweight_best_model.pth')

    return train_losses, val_losses

def evaluate_lightweight(model, test_loader, param_scaler, device):
    """Quick evaluation"""
    model.eval()
    predictions = []
    true_values = []

    with torch.no_grad():
        for features, targets in test_loader:
            features, targets = features.to(device), targets.to(device)
            outputs = model(features)

            outputs_cpu = outputs.cpu().numpy()
            targets_cpu = targets.cpu().numpy()

            if param_scaler:
                outputs_cpu = param_scaler.inverse_transform(outputs_cpu)
                targets_cpu = param_scaler.inverse_transform(targets_cpu)

            predictions.append(outputs_cpu)
            true_values.append(targets_cpu)

    predictions = np.concatenate(predictions, axis=0)
    true_values = np.concatenate(true_values, axis=0)

    # Calculate metrics
    mse = mean_squared_error(true_values, predictions)
    r2 = r2_score(true_values, predictions)

    param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']

    print(f"\n🎯 LIGHTWEIGHT MODEL RESULTS:")
    print(f"Overall MSE: {mse:.6f}")
    print(f"Overall R²: {r2:.6f}")
    print("-" * 40)

    for i, param in enumerate(param_names):
        param_mse = mean_squared_error(true_values[:, i], predictions[:, i])
        param_r2 = r2_score(true_values[:, i], predictions[:, i])
        print(f"{param:>6}: MSE={param_mse:>8.6f}, R²={param_r2:>6.6f}")

    return predictions, true_values, mse, r2

def transfer_learn_from_original(lightweight_model, original_model_path, device):
    """Transfer knowledge from the original larger model"""
    try:
        # Load original model state
        checkpoint = torch.load(original_model_path, map_location=device)

        # Transfer conv layers (they have similar structure)
        lightweight_state = lightweight_model.state_dict()

        # Map some layers from original to lightweight
        transfer_map = {
            'conv_layers.0.weight': 'conv_layers.0.weight',
            'conv_layers.0.bias': 'conv_layers.0.bias',
            'conv_layers.2.weight': 'conv_layers.2.weight',
            'conv_layers.2.bias': 'conv_layers.2.bias',
        }

        for orig_key, light_key in transfer_map.items():
            if orig_key in checkpoint and light_key in lightweight_state:
                # Handle size differences
                orig_param = checkpoint[orig_key]
                light_param = lightweight_state[light_key]

                if orig_param.shape == light_param.shape:
                    lightweight_state[light_key] = orig_param
                elif len(orig_param.shape) == len(light_param.shape):
                    # Truncate if original is larger
                    if all(o >= l for o, l in zip(orig_param.shape, light_param.shape)):
                        slices = tuple(slice(0, s) for s in light_param.shape)
                        lightweight_state[light_key] = orig_param[slices]

        lightweight_model.load_state_dict(lightweight_state)
        print("✅ Transferred knowledge from original model")

    except Exception as e:
        print(f"⚠️ Could not transfer from original model: {e}")
        print("Starting fresh training...")

# Main execution
if __name__ == "__main__":
    print("🚀 LIGHTWEIGHT FINE-TUNING FOR IZHIKEVICH NN")
    print("=" * 50)

    optimize_gpu_memory()

    # Setup device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Load dataset (assuming IzhikevichDataset is already defined)
    print("Loading dataset...")
    dataset = IzhikevichDataset("synthetic_dataset", sequence_length=500)  # Even shorter sequence for speed

    total_samples = len(dataset)
    print(f"Total dataset size: {total_samples}")

    # FIXED: Proper dataset splitting logic
    train_size = min(10000, int(0.7 * total_samples))  # Cap at 10k for speed
    val_size = min(2000, int(0.15 * total_samples))
    test_size = total_samples - train_size - val_size  # Use remaining samples

    # Ensure we don't have negative test_size
    if test_size < 500:  # Minimum test size
        # Adjust splits if dataset is too small
        train_size = int(0.7 * total_samples)
        val_size = int(0.15 * total_samples)
        test_size = total_samples - train_size - val_size

    print(f"Dataset split: Train={train_size}, Val={val_size}, Test={test_size}")
    print(f"Total check: {train_size + val_size + test_size} == {total_samples}")

    train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(
        dataset, [train_size, val_size, test_size],
        generator=torch.Generator().manual_seed(42)
    )

    # Larger batch sizes for speed
    batch_size = 128 if device.type == 'cuda' else 64

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,
                             num_workers=2, pin_memory=True if device.type == 'cuda' else False)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           num_workers=2, pin_memory=True if device.type == 'cuda' else False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,
                            num_workers=2, pin_memory=True if device.type == 'cuda' else False)

    # Create lightweight model
    model = LightweightIzhikevichNet(sequence_length=500, hidden_size=128, dropout=0.3)
    print(f"Lightweight model parameters: {sum(p.numel() for p in model.parameters()):,}")

    # Try to transfer from original model
    transfer_learn_from_original(model, 'regular_nn_model.pth', device)

    # Quick training
    print("\n🏃‍♂️ Starting fast training...")
    train_losses, val_losses = quick_train(
        model, train_loader, val_loader, device,
        epochs=15, lr=5e-4
    )

    # Load best model
    model.load_state_dict(torch.load('lightweight_best_model.pth', map_location=device))

    # Evaluate
    print("\n📊 Evaluating...")
    predictions, true_values, mse, r2 = evaluate_lightweight(
        model, test_loader, dataset.param_scaler if dataset.normalize else None, device
    )

    # Quick visualization
    print("\n📈 Creating quick visualization...")
    param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']

    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    axes = axes.flatten()

    for i, param in enumerate(param_names):
        ax = axes[i]
        ax.scatter(true_values[:, i], predictions[:, i], alpha=0.3, s=1)

        min_val = min(true_values[:, i].min(), predictions[:, i].min())
        max_val = max(true_values[:, i].max(), predictions[:, i].max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.8)

        ax.set_xlabel(f'True {param}')
        ax.set_ylabel(f'Predicted {param}')
        ax.set_title(f'{param}')

        r2_param = r2_score(true_values[:, i], predictions[:, i])
        ax.text(0.05, 0.95, f'R²={r2_param:.3f}', transform=ax.transAxes,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.savefig('lightweight_results.png', dpi=100, bbox_inches='tight')
    plt.show()

    # Training curve
    plt.figure(figsize=(8, 4))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Lightweight Training Progress')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.savefig('lightweight_training.png', dpi=100, bbox_inches='tight')
    plt.show()

    print(f"\n✅ LIGHTWEIGHT FINE-TUNING COMPLETE!")
    print(f"🎯 Final R²: {r2:.6f}")
    print(f"💾 Model saved as 'lightweight_best_model.pth'")
    print(f"📊 Results saved as 'lightweight_results.png'")
    print(f"⏱️ Training completed in ~3-5 minutes!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from tqdm import tqdm
import matplotlib.pyplot as plt

def optimize_gpu_memory():
    """Quick GPU optimization"""
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True
        print("GPU optimized")

class EnhancedLightweightIzhikevichNet(nn.Module):
    """Enhanced lightweight model with targeted improvements for problematic parameters"""

    def __init__(self, sequence_length, hidden_size=128, dropout=0.2):
        super(EnhancedLightweightIzhikevichNet, self).__init__()

        # Enhanced CNN layers with attention to problematic parameters
        self.conv_layers = nn.Sequential(
            nn.Conv1d(2, 16, kernel_size=7, padding=3),
            nn.ReLU(),
            nn.BatchNorm1d(16),
            nn.Conv1d(16, 32, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.MaxPool1d(2),
            nn.Conv1d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.Conv1d(64, 96, kernel_size=3, padding=1),  # Extra layer for complex patterns
            nn.ReLU(),
            nn.BatchNorm1d(96),
            nn.MaxPool1d(2),
            nn.Dropout1d(dropout),
        )

        # Bidirectional LSTM for better context
        self.lstm = nn.LSTM(96, hidden_size, 2, batch_first=True, dropout=dropout, bidirectional=True)
        lstm_output_size = hidden_size * 2  # Bidirectional

        # Shared feature extractor for parameter relationships
        self.shared_features = nn.Sequential(
            nn.Linear(lstm_output_size, 256),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(dropout/2)
        )

        # Specialized parameter heads with different architectures based on performance
        self.param_heads = nn.ModuleDict({
            # Good performers - keep simple
            'C': nn.Sequential(nn.Linear(128, 32), nn.ReLU(), nn.Linear(32, 1)),
            'k': nn.Sequential(nn.Linear(128, 32), nn.ReLU(), nn.Linear(32, 1)),
            'vr': nn.Sequential(nn.Linear(128, 32), nn.ReLU(), nn.Linear(32, 1)),
            'vt': nn.Sequential(nn.Linear(128, 32), nn.ReLU(), nn.Linear(32, 1)),
            'b': nn.Sequential(nn.Linear(128, 32), nn.ReLU(), nn.Linear(32, 1)),
            'd': nn.Sequential(nn.Linear(128, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1)),

            # Problematic parameters - more complex architectures
            'vpeak': nn.Sequential(
                nn.Linear(128, 128), nn.ReLU(), nn.Dropout(dropout),
                nn.Linear(128, 64), nn.ReLU(), nn.Dropout(dropout/2),
                nn.Linear(64, 32), nn.ReLU(),
                nn.Linear(32, 1)
            ),
            'a': nn.Sequential(
                nn.Linear(128, 96), nn.ReLU(), nn.Dropout(dropout),
                nn.Linear(96, 48), nn.ReLU(), nn.Dropout(dropout/2),
                nn.Linear(48, 1)
            ),
            'c': nn.Sequential(
                nn.Linear(128, 96), nn.ReLU(), nn.Dropout(dropout),
                nn.Linear(96, 48), nn.ReLU(), nn.Dropout(dropout/2),
                nn.Linear(48, 24), nn.ReLU(),
                nn.Linear(24, 1)
            ),
        })

        # Parameter interaction layer (optional ensemble)
        self.param_interaction = nn.MultiheadAttention(embed_dim=9, num_heads=3, dropout=dropout, batch_first=True)

    def forward(self, x):
        # CNN feature extraction
        conv_out = self.conv_layers(x)
        conv_out = conv_out.permute(0, 2, 1)

        # Bidirectional LSTM
        lstm_out, (h_n, c_n) = self.lstm(conv_out)
        # Use both forward and backward final states
        last_hidden = torch.cat([h_n[-2], h_n[-1]], dim=1)  # Concatenate final forward and backward

        # Shared features
        shared_feat = self.shared_features(last_hidden)

        # Parameter-specific predictions
        outputs = []
        param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']
        for param in param_names:
            outputs.append(self.param_heads[param](shared_feat))

        param_tensor = torch.cat(outputs, dim=1)

        # Optional: Parameter interaction refinement
        param_tensor_expanded = param_tensor.unsqueeze(1)  # Add sequence dimension
        refined_params, _ = self.param_interaction(param_tensor_expanded, param_tensor_expanded, param_tensor_expanded)
        refined_params = refined_params.squeeze(1)

        # Combine original and refined predictions
        final_output = 0.7 * param_tensor + 0.3 * refined_params

        return final_output

def adaptive_weighted_mse_loss(predictions, targets, epoch=0, max_epochs=15):
    """Adaptive weighted loss that adjusts based on training progress"""
    # Base weights for problematic parameters
    base_weights = torch.tensor([1.0, 1.0, 1.0, 1.0, 5.0, 3.0, 1.0, 3.0, 1.0])  # vpeak=5, a=3, c=3

    # Adaptive scaling - higher weights early in training
    adaptive_scale = 1.0 + (max_epochs - epoch) / max_epochs * 2.0  # Scale from 3.0 to 1.0
    problematic_mask = torch.tensor([0, 0, 0, 0, 1, 1, 0, 1, 0])  # vpeak, a, c
    weights = base_weights + problematic_mask * adaptive_scale

    weights = weights.to(predictions.device)
    mse_per_param = torch.mean((predictions - targets) ** 2, dim=0)
    weighted_loss = torch.sum(weights * mse_per_param)
    return weighted_loss

def enhanced_train(model, train_loader, val_loader, device, epochs=20, base_lr=3e-4):
    """Enhanced training with adaptive strategies"""
    model.to(device)

    # Separate optimizers for different parts with different learning rates
    conv_params = list(model.conv_layers.parameters())
    lstm_params = list(model.lstm.parameters()) + list(model.shared_features.parameters())
    head_params = list(model.param_heads.parameters())
    attention_params = list(model.param_interaction.parameters())

    optimizer = optim.AdamW([
        {'params': conv_params, 'lr': base_lr * 0.3, 'weight_decay': 1e-4},
        {'params': lstm_params, 'lr': base_lr * 0.8, 'weight_decay': 1e-5},
        {'params': head_params, 'lr': base_lr * 1.5, 'weight_decay': 1e-5},
        {'params': attention_params, 'lr': base_lr * 0.5, 'weight_decay': 1e-4}
    ])

    # Cosine annealing with warm restarts
    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=5, T_mult=2, eta_min=base_lr*0.01)

    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    patience = 7
    patience_counter = 0

    for epoch in range(epochs):
        # Training
        model.train()
        train_loss = 0.0

        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for batch_idx, (features, targets) in enumerate(pbar):
            features, targets = features.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = model(features)

            # Adaptive weighted loss
            loss = adaptive_weighted_mse_loss(outputs, targets, epoch, epochs)
            loss.backward()

            # Gradient clipping with different norms for different parts
            torch.nn.utils.clip_grad_norm_(conv_params, max_norm=0.5)
            torch.nn.utils.clip_grad_norm_(lstm_params, max_norm=1.0)
            torch.nn.utils.clip_grad_norm_(head_params, max_norm=2.0)
            torch.nn.utils.clip_grad_norm_(attention_params, max_norm=0.3)

            optimizer.step()
            scheduler.step(epoch + batch_idx / len(train_loader))

            train_loss += loss.item()
            pbar.set_postfix({'loss': f'{loss.item():.4f}', 'lr': f'{optimizer.param_groups[0]["lr"]:.6f}'})

        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for features, targets in val_loader:
                features, targets = features.to(device), targets.to(device)
                outputs = model(features)
                loss = adaptive_weighted_mse_loss(outputs, targets, epoch, epochs)
                val_loss += loss.item()

        train_loss /= len(train_loader)
        val_loss /= len(val_loader)

        train_losses.append(train_loss)
        val_losses.append(val_loss)

        print(f"Epoch {epoch+1}: Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

        # Early stopping and model saving
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'enhanced_lightweight_best_model.pth')
            print(f"✅ New best model saved (Val Loss: {val_loss:.4f})")
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f"⏰ Early stopping at epoch {epoch+1}")
            break

    return train_losses, val_losses

def detailed_evaluation(model, test_loader, param_scaler, device):
    """Detailed evaluation with parameter-specific analysis"""
    model.eval()
    predictions = []
    true_values = []

    with torch.no_grad():
        for features, targets in test_loader:
            features, targets = features.to(device), targets.to(device)
            outputs = model(features)

            outputs_cpu = outputs.cpu().numpy()
            targets_cpu = targets.cpu().numpy()

            if param_scaler:
                outputs_cpu = param_scaler.inverse_transform(outputs_cpu)
                targets_cpu = param_scaler.inverse_transform(targets_cpu)

            predictions.append(outputs_cpu)
            true_values.append(targets_cpu)

    predictions = np.concatenate(predictions, axis=0)
    true_values = np.concatenate(true_values, axis=0)

    # Calculate metrics
    mse = mean_squared_error(true_values, predictions)
    r2 = r2_score(true_values, predictions)

    param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']

    print(f"\n🎯 ENHANCED LIGHTWEIGHT MODEL RESULTS:")
    print(f"Overall MSE: {mse:.6f}")
    print(f"Overall R²: {r2:.6f}")
    print("=" * 50)

    improvements = []
    for i, param in enumerate(param_names):
        param_mse = mean_squared_error(true_values[:, i], predictions[:, i])
        param_r2 = r2_score(true_values[:, i], predictions[:, i])

        # Color coding based on performance
        if param_r2 > 0.8:
            status = "🟢 EXCELLENT"
        elif param_r2 > 0.65:
            status = "🟡 GOOD    "
        else:
            status = "🔴 NEEDS WORK"

        print(f"{param:>6}: MSE={param_mse:>8.6f}, R²={param_r2:>6.6f} {status}")

        # Track improvements for problematic parameters
        if param in ['vpeak', 'a', 'c']:
            improvements.append((param, param_r2))

    print("=" * 50)
    print("🔍 FOCUS AREAS IMPROVEMENT:")
    original_scores = {'vpeak': 0.251, 'a': 0.605, 'c': 0.589}  # From previous results
    for param, new_score in improvements:
        old_score = original_scores[param]
        improvement = new_score - old_score
        print(f"{param}: {old_score:.3f} → {new_score:.3f} (Δ{improvement:+.3f})")

    return predictions, true_values, mse, r2

# Main execution for enhanced model
if __name__ == "__main__":
    print("🚀 ENHANCED LIGHTWEIGHT MODEL FOR IZHIKEVICH NN")
    print("=" * 55)

    optimize_gpu_memory()

    # Setup device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Load dataset
    print("Loading dataset...")
    dataset = IzhikevichDataset("synthetic_dataset", sequence_length=500)

    total_samples = len(dataset)
    print(f"Total dataset size: {total_samples}")

    # Dataset splitting
    train_size = min(12000, int(0.7 * total_samples))  # Slightly larger for better training
    val_size = min(2500, int(0.15 * total_samples))
    test_size = total_samples - train_size - val_size

    print(f"Dataset split: Train={train_size}, Val={val_size}, Test={test_size}")

    train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(
        dataset, [train_size, val_size, test_size],
        generator=torch.Generator().manual_seed(42)
    )

    # Data loaders
    batch_size = 96 if device.type == 'cuda' else 48  # Optimized batch size

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,
                             num_workers=2, pin_memory=True if device.type == 'cuda' else False)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           num_workers=2, pin_memory=True if device.type == 'cuda' else False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,
                            num_workers=2, pin_memory=True if device.type == 'cuda' else False)

    # Create enhanced model
    model = EnhancedLightweightIzhikevichNet(sequence_length=500, hidden_size=128, dropout=0.2)
    total_params = sum(p.numel() for p in model.parameters())
    print(f"Enhanced model parameters: {total_params:,}")

    # Enhanced training
    print("\n🏃‍♂️ Starting enhanced training...")
    train_losses, val_losses = enhanced_train(
        model, train_loader, val_loader, device,
        epochs=70, base_lr=3e-4
    )

    # Load best model
    model.load_state_dict(torch.load('enhanced_lightweight_best_model.pth', map_location=device))

    # Detailed evaluation
    print("\n📊 Detailed evaluation...")
    predictions, true_values, mse, r2 = detailed_evaluation(
        model, test_loader, dataset.param_scaler if dataset.normalize else None, device
    )

    # Enhanced visualization
    print("\n📈 Creating enhanced visualization...")
    param_names = ['C', 'k', 'vr', 'vt', 'vpeak', 'a', 'b', 'c', 'd']

    fig, axes = plt.subplots(3, 3, figsize=(15, 15))
    axes = axes.flatten()

    for i, param in enumerate(param_names):
        ax = axes[i]

        # Color points based on error magnitude
        errors = np.abs(true_values[:, i] - predictions[:, i])
        scatter = ax.scatter(true_values[:, i], predictions[:, i],
                           c=errors, cmap='viridis', alpha=0.6, s=2)

        min_val = min(true_values[:, i].min(), predictions[:, i].min())
        max_val = max(true_values[:, i].max(), predictions[:, i].max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.8, linewidth=2)

        ax.set_xlabel(f'True {param}')
        ax.set_ylabel(f'Predicted {param}')
        ax.set_title(f'{param} - Enhanced Model')

        r2_param = r2_score(true_values[:, i], predictions[:, i])
        mse_param = mean_squared_error(true_values[:, i], predictions[:, i])

        # Color-coded text box
        if r2_param > 0.8:
            bbox_color = 'lightgreen'
        elif r2_param > 0.65:
            bbox_color = 'lightyellow'
        else:
            bbox_color = 'lightcoral'

        ax.text(0.05, 0.95, f'R²={r2_param:.3f}\nMSE={mse_param:.3f}',
                transform=ax.transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor=bbox_color, alpha=0.8))

        plt.colorbar(scatter, ax=ax, shrink=0.6)

    plt.tight_layout()
    plt.savefig('enhanced_lightweight_results.png', dpi=150, bbox_inches='tight')
    plt.show()

    # Training progress
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss', linewidth=2)
    plt.plot(val_losses, label='Val Loss', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Enhanced Training Progress')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.subplot(1, 2, 2)
    # Plot parameter-specific improvements
    original_r2 = [0.820, 0.730, 0.935, 0.687, 0.251, 0.605, 0.856, 0.589, 0.774]
    current_r2 = [r2_score(true_values[:, i], predictions[:, i]) for i in range(9)]

    x = np.arange(9)
    width = 0.35
    plt.bar(x - width/2, original_r2, width, label='Original', alpha=0.7)
    plt.bar(x + width/2, current_r2, width, label='Enhanced', alpha=0.7)
    plt.xlabel('Parameters')
    plt.ylabel('R² Score')
    plt.title('Parameter Performance Comparison')
    plt.xticks(x, param_names)
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('enhanced_training_comparison.png', dpi=150, bbox_inches='tight')
    plt.show()

    print(f"\n✅ ENHANCED LIGHTWEIGHT TRAINING COMPLETE!")
    print(f"🎯 Final R²: {r2:.6f}")
    print(f"💾 Model saved as 'enhanced_lightweight_best_model.pth'")
    print(f"📊 Results saved as 'enhanced_lightweight_results.png'")
    print(f"⚡ Enhanced model with {total_params:,} parameters!")
```
